```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import *
from scipy import *
```

Finding 1: Gun control can't stop suicide. It only changes the methods of suicide.

```python
#Data for finding1:
df = pd.read_csv('suicides_aus.csv')
df.head(5)
```

```python
#define variables
year = df["year"]
data1 = df["suicides number"]
data2 = df["suicides by firearm"]
```

```python
#visualize it with line graph since we want to contrast the difference.

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('year')
ax1.set_ylabel('Australia suicides number', color=color)
ax1.plot(year, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('Australia suicides by firearm', color=color)  # we already handled the x-label with ax1
ax2.plot(year, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

#add title
plt.title("Gun control can't stop suicide.\nIt only changes the methods of suicide.", fontsize=16)

#add a verticle line to indicate the timing of gun buyback program
plt.axvline(x="1996", color='black', linestyle='--')

#add data source
plt.annotate('Data source: ABS', (0,0), (-60,-20), fontsize=8, 
             xycoords='axes fraction', textcoords='offset points', va='top')

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```

Finding3: More guns, less crime.

```python
#Data for finding3:
df3 = pd.read_csv('guns.csv')
df3.head(5)
```

```python
#define variables
year = df3["Year"]
data1 = df3["guns per person"]
data2 = df3["Firearm Homicide Deaths Rate"]
```

```python
#visualize it with line graph since we want to contrast the difference.

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('year')
ax1.set_ylabel('guns per person in the US', color=color)
ax1.plot(year, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('US Firearm Homicide Deaths Rate', color=color)  # we already handled the x-label with ax1
ax2.plot(year, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

#add title
plt.title("More guns, less crime.", fontsize=16)

#add data source
plt.annotate('Data source: ATF', (0,0), (-60,-20), fontsize=8, 
             xycoords='axes fraction', textcoords='offset points', va='top')

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.show()
```
